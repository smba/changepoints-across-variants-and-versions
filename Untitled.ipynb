```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import itertools

plt.style.use('ggplot')
```

```python
def make_model(ys: np.ndarray, obs: np.ndarray, threshold: float = 0.01, exp = 2):
    count = np.zeros(ys.shape[0])
    for a, b in itertools.combinations(obs, 2):
        delta = np.abs(ys[a] - ys[b])
        dist = np.abs(a - b)
        if delta > threshold * ys[a] or delta > threshold * ys[b]:
            count[min(a, b):max(a, b)] += 1/dist**exp
    return count/np.sum(count)

ys = np.ones(1000)
ys[:300] += 1
ys[500:] += 4

obs = np.random.choice(np.arange(0,1000), size=10).tolist()

count = make_model(ys, obs, exp=1)
plt.plot(count); plt.fill_between(np.arange(1000), np.zeros(1000), count, alpha=0.5)
plt.show()

for i in range(10):
    unseen = list(set(np.arange(1000)) - set(obs))
    unseen = np.arange(1000)[unseen]
    p = count[unseen]/np.sum(count[unseen])
    
    new_ons = np.random.choice(unseen, p=p, size=3).tolist()
    
    obs = obs + new_ons
    
    count2 = make_model(ys, obs)
    
    # get medians
    median = np.median(count2)
    print(np.argwhere((count2 < median*1.001) & (count2 > median * 0.999)))
    
    #plt.title("{} %".format(len(obs)/1000*100))
    #plt.plot(count2); plt.fill_between(np.arange(1000), np.zeros(1000), count2,  alpha=0.5)
    #plt.show()
    
```
